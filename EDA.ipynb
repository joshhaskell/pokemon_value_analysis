# Clean Data

## Deal with Missing Values

```python
clean_df = df.copy()

# Fill values with 'Unknown'
fill_unknowns_columns = ['rarity','artist','evolvesFrom','regulationMark','set_legalities_expanded','set_legalities_standard',
                   'ancientTrait_name','type_1','type_2','subtype_1','subtype_2','subtype_3','subtype_4','nationalPokedexNumber']
# Fill values with 0
fill_zeros = ['hp','convertedRetreatCost', 'num_abilities', 'num_attacks', 'num_resistances', 'num_weaknesses']

clean_df[fill_unknowns_columns] = clean_df[fill_unknowns_columns].fillna('Unknown')
clean_df[fill_zeros] = clean_df[fill_zeros].fillna(0)
```

For any missing market prices, will fill in with mid, then average of low and high, then low, than high prices. There were only 85 that could be filled so for the rest, there are not prices at all. Will drop the price_type, low, mid, high and directLow prices and use just the market price moving forward. 

After filling prices

```python
# Prices
print(f"Missing 'market' prices before filling: {clean_df['market'].isnull().sum()}")

# Fill missing 'market' prices using the given strategy
clean_df['market'] = (clean_df['market'].fillna(clean_df['mid'])
                           .fillna(clean_df[['low', 'high']].mean(axis=1))
                           .fillna(clean_df['low'])
                           .fillna(clean_df['high']))
print(f"Missing 'market' prices after filling: {clean_df['market'].isnull().sum()}")

# Fill remaining NA values with the median of the 'market' column within each 'price_type' group
clean_df['market'] = clean_df.groupby('price_type')['market'].transform(lambda x: x.fillna(x.median()))

# For those with unknown price type fill values with the overall median of the 'market' column
clean_df['market'] = clean_df['market'].fillna(clean_df['market'].median())

print(f"Missing 'market' prices after filling with median by price_type: {clean_df['market'].isnull().sum()}")

# Create price_bucket column
clean_df['price_bucket'] = clean_df['market'].apply(lambda x: 'low' if x < 50 else 'high')
```

```python
# Year
# Convert set_releaseDate to datetime
clean_df['set_releaseDate'] = pd.to_datetime(clean_df['set_releaseDate'], format='%Y/%m/%d', errors='coerce')
# Extract year from set_releaseDate
clean_df['release_year'] = clean_df['set_releaseDate'].dt.year
```

## Drop Columns After Visualization

### Evaluation of those filled with 'Unknown'

```python
# Initialize an empty list to store the results
results = []

# List of columns that were originally filled with 'Unknown'
original_na_columns = clean_df.columns[clean_df.isin(['Unknown']).any()]

for column in original_na_columns:
    # Calculate the original proportion of 'Unknown' values
    unknown_count = (clean_df[column] == 'Unknown').sum()
    unknown_proportion = (unknown_count / len(clean_df) * 100).round(2)
    
    # Calculate the proportion of 'high' price buckets within the filled values
    filled_values = clean_df[clean_df[column] != 'Unknown']
    high_price_proportion = ((filled_values['price_bucket'] == 'high').mean() * 100).round(2)
    
    # Append the results to the list
    results.append({
        'Column': column,
        'Proportion of Unknown Values (%)': unknown_proportion,
        'Proportion of High Price Bucket in Filled Values (%)': high_price_proportion
    })

# Convert the results list to a DataFrame
results_df = pd.DataFrame(results)
```

```python
# Dropping columns with more than 30% unknown values
results_df.sort_values('Proportion of High Price Bucket in Filled Values (%)', ascending=False)
```

```python
# Dropping columns with more than 30% unknown values since they did not seem to have a significant impact on the price bucket
columns_to_drop = results_df[results_df['Proportion of Unknown Values (%)'] > 30]['Column'].unique()
clean_df = clean_df.drop(columns=columns_to_drop, axis=1)
```
